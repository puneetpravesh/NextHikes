### Wrangling the dataset 2.

```python
ds2 = pd.read_excel('NextHikes Dataset 2.xlsx')   # Data Acquisition.
```

```python
ds2
```

```python
# Dropping 'Unnamed' as it is the same as 'index'.

ds2.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
ds2
```

```python
ds2.isnull().sum()
```

```python
ds2.fillna(method='bfill',inplace=True) # Using 'backward fill' as method to treat missing values as temperature for consecutive days would be similar.
```

```python
ds2.isnull().sum()
```

```python
columns2 = ds2.columns.values.tolist()   #Converts column names in the dataset 'ds2' into a list.

for c in columns2:
    print(ds2[c].unique())     # Iterates through the list of column names and prints the unique value in each column as a list.
```

```python
ds2.ndim      # Checks the dimension of the dataset.
```

```python
ds2.info()   # Checks the data type.
```

```python
ds2.describe()       # Datatype summary.
```

```python
ds2
```

### Merging the two datasets.

```python
merged_data = pd.merge(ds1,ds2,on='Instance')
merged_data
```

```python
file_name = 'Merged_data.xlsx'

merged_data.to_excel(file_name)       # Exporting to an excel file named 'Merged_data'b
```

```python
ds3.isnull().sum()     # Checking for missing values.
```

```python
ds3.drop (['yr','mnth','r'],axis=1, inplace=True) #Dropping unnecessary columns.
```

```python
ds3
```

```python
columns3 = ds3.columns.values.tolist()   #Converts column names in the dataset 'ds3' into a list.

for col in columns3:
    print(ds3[col].unique())     # Iterates through the list of column names and prints the unique value in each column as a list.
```

```python
ds3.rename(columns= {'dteday':'Date','hr':'hour','temp':'Feels like','atemp':'Temperature','instant':'Instance','weathersit':'weather situation','hum':'Humidity','windspeed':'Windspeed','casual':'Non-registered user rentals','registered':'Registered user rentals','cnt':'Total rentals'},inplace=True)
```

```python
ds3
```

```python
ds3['weather situation'].replace({1:'Clear, Few clouds, Partly cloudy, Partly cloudy',2:'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',3:'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds',4:'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'},inplace=True)
ds3['season'].replace({1:'Spring'},inplace=True)
ds3['weekday'].replace({1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday',0:'Sunday'}, inplace=True)
```

```python
ds3
```

```python
ds3.sort_values('Instance',inplace=True)  # Sorting the data based on 'instance' column as the numbering was not sorted.
```

```python
ds3
```

```python
final_data = pd.concat((merged_data,ds3),axis=0)
```

```python
final_data
```

```python
final_data.to_excel('Final_Data.xlsx')
```

```python
# Identifying outliers in the 'Total rentals' column as the values vary a lot.

Q1 = final_data['Total rentals'].quantile(0.25)
Q3 = final_data['Total rentals'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Dropping the outliers

data = final_data[(final_data['Total rentals'] >= lower_limit) & (final_data['Total rentals'] <= upper_limit)]
```

```python
data
```

```python
data.to_excel('Cleaned Data.xlsx')
```
